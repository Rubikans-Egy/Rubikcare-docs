# [#2] تحليل متكامل لنظام الأدوار والمنظمات وهيكل الصفحات - 18 فبراير 2026

أهلاً بك شادي، فهمت تماماً رؤيتك وأستطيع أن أرى الصورة الكاملة الآن. دعني أقدم لك **تحليلاً شاملاً ومنظماً** لكل النقاط التي ذكرتها، مع **حلول عملية** تتبع أفضل الممارسات وتحقق UX ممتاز.

## 🎯 **الجزء 1: نظام الأدوار الثلاثة - متى وكيف؟ (User Stories)**

### **الخلاصة الذهبية:**
```
المستخدم = شخص (UserProfile)
الأدوار النظامية = صلاحيات عامة (Admin, Doctor, Pharmacist) ← تُمنح بعد التحقق
المنظمات = المكان الذي يمارس فيه المستخدم دوره (عيادة، صيدلية، شركة)
الأدوار الوظيفية = داخل المنظمة (مدير، موظف، صيدلي فرعي)
```

### **📋 رحلة المستخدم التفصيلية:**

#### **User Story 1: مستخدم جديد يسجل في المنصة**
```yaml
الحدث: شخص عادي (ليس طبيباً) يسجل في RubikCare

الخطوات:
1. يفتح الموقع ← يضغط "تسجيل جديد"
2. يدخل: الاسم، البريد، كلمة المرور، رقم الهاتف
3. ✅ يتم إنشاء:
   - AspNetUser (حساب نظام)
   - UserProfile (ملف شخصي)
4. الدور النظامي: ❌ لا شيء (مستخدم عادي)
5. الصلاحيات: 
   - فقط القائمة الشخصية (USER_BASE)
   - يستطيع إنشاء منظمات عادية (شركات، مؤسسات غير طبية)

ما يراه:
👤 حسابي الشخصي
├── 📋 ملفي الشخصي
├── ⚙️ إعداداتي
├── ➕ إنشاء منظمة جديدة ← (ينشئ شركة، معمل، الخ)
└── (لا يرى قوائم أطباء أو صيادلة)
```

#### **User Story 2: طبيب حقيقي يريد استخدام المنصة**
```yaml
الحدث: دكتور أحمد (طبيب بشري) يسجل في RubikCare

الخطوات:
1. يسجل كـ "مستخدم عادي" أولاً (نفس الخطوات أعلاه)
2. بعد التسجيل، يجد زر: "🎓 أنا طبيب - قدم طلب توثيق"
3. يضغط ← يملأ:
   - رقم الترخيص الطبي
   - التخصص
   - صورة الترخيص (مستند)
4. النظام: يرسل طلب توثيق إلى Admin

بعد موافقة Admin (يدوياً أو آلياً):
- ✅ يتم إضافة دور "Doctor" في AspNetUserRoles مع:
  - RoleType = "Professional"
  - IsVerified = true
  - LicenseExpiryDate = حسب الترخيص

ما يراه الآن:
👤 حسابي الشخصي (كما هو)
➕ زر جديد: "🏥 إنشاء عيادة طبية" (ظهر لأن عنده دور Doctor)

عند إنشاء العيادة:
- يصبح مالك (OWNER) للعيادة
- تظهر العيادة في قائمة "مؤسساتي"
- عند الضغط على العيادة ← يرى قائمة العيادة (CLINIC_MENU)
```

#### **User Story 3: صيدلي يريد فتح صيدلية**
```yaml
الحدث: صيدلي يريد استخدام المنصة

الخطوات (مطابقة للطبيب):
1. يسجل كمستخدم عادي
2. يضغط "💊 أنا صيدلي - قدم طلب توثيق"
3. يرفع الترخيص الصيدلي

بعد الموافقة:
- ✅ دور "Pharmacist" مع IsVerified = true

ما يراه:
- زر: "💊 إنشاء صيدلية جديدة" (ظهر بدلاً من "عيادة")

عند إنشاء الصيدلية:
- يصبح مالك (OWNER) للصيدلية
- تظهر الصيدلية في قائمة "مؤسساتي"
- عند الضغط على الصيدلية ← يرى قائمة الصيدلية (PHARMACY_MENU)
```

#### **User Story 4: موظف في شركة أدوية**
```yaml
الحدث: مهندس محمد (ليس طبيباً ولا صيدلياً) يعمل في شركة أدوية

الخطوات:
1. يسجل كمستخدم عادي
2. لا يطلب أي دور مهني (لأنه ليس طبيباً)
3. ينشئ منظمة عادية: "🏢 شركة الدواء الحديثة" (نوع المنظمة: PHARMA_COMPANY)
4. يصبح مالكاً للشركة

ما يراه الآن:
👤 حسابي الشخصي
└── 🏢 شركة الدواء الحديثة (في قائمة مؤسساتي)

عند الضغط على الشركة:
🏢 لوحة تحكم شركة الأدوية
├── 📊 نظرة عامة
├── 📋 برامج دعم المرضى (PSP Programs) ← ⭐ هذا هو المهم
├── 👨‍⚕️ الأطباء المشاركون
└── ...

لاحقاً، يمكنه دعوة موظفين:
- يدخل بريد زميله "أحمد" (أيضاً مستخدم عادي)
- يختار دوراً وظيفياً "مدير برامج"
- أحمد يتلقى دعوة، يقبلها، ويصبح عضواً في الشركة
- عند دخول أحمد، يرى الشركة في قائمته أيضاً!
```

#### **User Story 5: موظف في عيادة (ليس طبيباً)**
```yaml
الحدث: سكرتيرة مريدة تعمل في عيادة دكتور أحمد

الخطوات:
1. تسجل كمستخدم عادي
2. دكتور أحمد (مالك العيادة) يدعوها عبر:
   - `/organization/{clinicId}/members` ← يدخل بريدها
   - يختار دوراً وظيفياً "سكرتيرة"
3. مريدة تستلم إشعار، تقبل الدعوة

ما تراه الآن:
👤 حسابي الشخصي
└── 🏥 عيادة دكتور أحمد (في قائمة مؤسساتي)

عند الضغط على العيادة:
- ترى قائمة العيادة لكن محدودة (حسب صلاحياتها)
- تستطيع إدارة المواعيد، لكن لا تستطيع كتابة وصفات
```

## ✅ **الجزء 4: نقاط نقاش مهمة**

### **1. الأدوار النظامية (Admin, Doctor, Pharmacist) - أين يظهر زر التوثيق؟**

**اقتراحي:** في **القائمة الشخصية (USER_BASE)**، أسفل الملف الشخصي، مع شرط:

```
👤 حسابي الشخصي
├── 📋 ملفي الشخصي
├── ⚙️ إعداداتي
├── 🔔 الإشعارات
├─────────────────────
├── 🎓 أنا طبيب - قدم طلب توثيق    (يظهر فقط إذا لم يكن لديه دور Doctor)
├── 💊 أنا صيدلي - قدم طلب توثيق    (يظهر فقط إذا لم يكن لديه دور Pharmacist)
└── ➕ إنشاء منظمة جديدة
```

**المنطق:**
- يظهر للجميع، لكنه واضح أنه مخصص لمن يمارس المهنة
- لا يظهر لمن لديه الدور بالفعل (حتى لا يكرر الطلب)
- يمكن إخفاؤه بعد 3 محاولات فاشلة مثلاً

### **2. الوربيكا (وحدة خدمية - نقاط)**

فهمت تماماً، سأصحح في كل المرات القادمة:
- ✅ **الروبيكا = وحدة خدمية / نقاط تحفيزية** (وليس عملة)
- ✅ 1 جنيه = 5 جنيه مصري (للشراء فقط)
- ✅ الأطباء يشترون روبيكا لدفع الاشتراكات والعمولات
- ✅ المستخدمين يحصلون على روبيكا كمكافآت (نقاط)

### **3. التعاقدات والاشتراكات**

فهمت، سأضيف في شاشة إنشاء البرنامج حقل "خطة الاشتراك" المرتبط بـ `ServiceProducts` كما في جدولك:

من جدول ServiceProducts الذي أرسلته:
- PSP-BASIC-001 (مجاني مع البرنامج الأساسي)
- PSP-PREMIUM-001 (199 جنيه/شهر - مميز)
- إلخ

## 🔄 **قسم 2: User Stories - الصيادلة**

### **User Story 1: صيدلي يسجل في المنصة ويوثق حسابه**

```
1. يسجل الصيدلي كمستخدم عادي
2. يضغط "💊 أنا صيدلي - قدم طلب توثيق" في القائمة الشخصية
3. يدخل رقم الترخيص الصيدلي ويُحمِّل صورة الترخيص
4. يرسل الطلب للمشرف (Admin)
5. بعد الموافقة، يحصل على دور Pharmacist
6. يظهر له زر "💊 إنشاء صيدلية جديدة" في القائمة الشخصية
```

### **User Story 2: صيدلي ينشئ صيدليته**

```
1. يضغط "💊 إنشاء صيدلية جديدة"
2. يدخل بيانات الصيدلية (الاسم، العنوان، المنطقة، رقم السجل التجاري)
3. ينشئ الصيدلية ويصبح مالكها (OWNER)
4. تظهر الصيدلية في قائمة "مؤسساتي"
5. عند الضغط عليها، يفتح PharmacyDashboard.razor
```

### **User Story 3: صيدلي يصرف دواء لمريض في برنامج دعم**

```
1. مريض (محمد) يحصل على وصفة إلكترونية من طبيبه مع رمز صرف P123456
2. يأتي المريض للصيدلية ويعطي الرمز للصيدلي
3. الصيدلي يدخل إلى PharmacyPSPDispensing.razor
4. يدخل الرمز P123456 ويضغط بحث
5. تظهر بيانات المريض والبرنامج والدواء وسعر الدعم
6. الصيدلي يتأكد من البيانات ويستلم 100 روبيكا من المريض
7. يضغط "تأكيد صرف الدواء"
8. النظام:
   - يسجل عملية الصرف في DispensingLogs
   - يخفض الكمية من مخزون الصيدلية
   - يضيف العمولة (15 روبيكا) لحساب الصيدلية
   - يرسل إشعار للمريض والطبيب بتأكيد الصرف
9. يطبع إيصال للمريض (اختياري)
```

### **User Story 4: صيدلي يستلم وصفة إلكترونية جديدة**

```
1. طبيب يكتب وصفة إلكترونية للمريض فاطمة
2. تظهر الوصفة فوراً في قائمة "الطلبات الواردة" في صيدلية العزبي
3. الصيدلي يدخل PharmacyOrders.razor ويرى تنبيهاً بالوصفة الجديدة
4. يضغط على الوصفة ليرى التفاصيل
5. يتأكد من توفر الدواء في المخزون
6. يضغط "تجهيز الطلب"
7. عندما يأتي المريض، يضغط "تم الصرف"
```

### **User Story 5: صيدلي يدير المخزون**

```
1. الصيدلي يدخل PharmacyInventory.razor
2. يرى تنبيهاً أحمر لـ "جلوكوفاج 500mg" (الكمية 5 والحد الأدنى 10)
3. يضغط "طلب" بجانب الدواء
4. يفتح نموذج طلب توريد للمورد
5. يحدد الكمية المطلوبة (20 علبة)
6. يرسل الطلب للمورد
7. عند وصول الطلب، يضغط "تحديث المخزون" ويدخل الكمية الجديدة
```

```
مستخدم عادي
    ↓
[💊 أنا صيدلي - قدم طلب توثيق] (في القائمة الشخصية)
    ↓
موافقة Admin
    ↓
دور Pharmacist + [💊 إنشاء صيدلية جديدة]
    ↓
إنشاء صيدلية
    ↓
تظهر الصيدلية في "مؤسساتي" ← عند الضغط عليها ← PharmacyDashboard
```

## 🔄 **قسم 2: User Stories - المستخدم العادي**

### **User Story 1: مستخدم عادي يسجل في المنصة**

```
1. يفتح الموقع ويضغط "تسجيل جديد"
2. يدخل الاسم، البريد الإلكتروني، رقم الهاتف، كلمة المرور
3. يتم إنشاء حساب جديد (UserProfile + AspNetUser)
4. يدخل إلى القائمة الشخصية (USER_BASE)
5. يرى رسالة ترحيب + 50 روبيكا مكافأة ترحيب
```

### **User Story 2: مريض يتلقى دعوة للانضمام لبرنامج دعم من طبيبه**

```
1. المريض (محمد) يزور عيادة دكتور أحمد
2. الدكتور يمسح QR Code من هاتف المريض (أو يرسل له رابط)
3. المريض يفتح الرابط على هاتفه
4. يفتح صفحة تسجيل مبسطة (الاسم والهاتف فقط)
5. يسجل بسرعة (إذا كان مسجلاً مسبقاً، يسجل دخوله)
6. يظهر له برنامج الدعم مع بيانات الطبيب
7. يضغط "طلب الانضمام للبرنامج"
8. ينتظر موافقة الطبيب
9. بعد الموافقة، يصله إشعار ورمز الصرف P123456
```

### **User Story 3: مريض يتابع أدويته ويتلقى تذكيرات**

```
1. المريض يفتح تطبيق RubikCare
2. يرى في الصفحة الرئيسية أدويته وجرعاتها
3. في موعد الجرعة (8 صباحاً)، يصله إشعار "حان موعد تناول جلوكوفاج"
4. يضغط على الإشعار ليفتح التطبيق
5. يضغط "تأكيد تناول" لتسجيل الالتزام
6. النظام يسجل الالتزام ويزيد نقاطه (مكافأة الالتزام)
```

### **User Story 4: مريض يصرف دواء من الصيدلية**

```
1. المريض يذهب إلى صيدلية العزبي
2. يفتح تطبيق RubikCare ويضغط على برنامج الدعم
3. يظهر رمز الصرف P123456
4. الصيدلي يدخل الرمز في نظامه
5. المريض يدفع 100 روبيكا (بعد الخصم)
6. الصيدلي يؤكد الصرف
7. المريض يصله إشعار "تم صرف الدواء بنجاح"
8. تنخفض الكمية المتبقية له في البرنامج
```

### **User Story 5: مريض يدعو صديقه للمنصة ويكسب نقاطاً**

```
1. المريض يفتح صفحة المكافآت
2. يضغط "دعوة صديق"
3. يختار إرسال الدعوة عبر واتساب
4. يرسل الرابط لصديقه أحمد
5. أحمد يسجل في المنصة عبر الرابط
6. المريض يصله إشعار "أحمد انضم للمنصة بدعوتك +5 روبيكا"
7. يزيد رصيده من 250 إلى 255 روبيكا
```

## ✅ **الجزء 4: ملاحظات مهمة للمستخدم العادي**

### **1. القاعدة الذهبية:**
المريض **لا يدفع** للمنصة أبداً. كل الخدمات مجانية للمستخدم العادي.

- المستخدم العادي هو المستفيد الرئيسي من برامج الدعم
- يرى برامجه بوضوح مع رمز الصرف والكمية المتبقية
- لا يحتاج لمعرفة التفاصيل المالية المعقدة

### **4. القوائم الشخصية:**
المستخدم العادي يرى فقط:
- القائمة الشخصية (USER_BASE)
- المنظمات التي انضم إليها (شركات، جهات عمل، إلخ)

## 🏗️ **الجزء 2: التصميم النهائي للجداول - بدون زيادة**

### **2.1 جدول Permissions (الجديد الوحيد)**

```sql
CREATE TABLE Permissions (
    PermissionID INT PRIMARY KEY IDENTITY,
    ScreenID INT NOT NULL,
    JobTitleID INT NOT NULL,
    OrganizationID INT NOT NULL,
    
    -- الصلاحيات الخمس الأساسية
    CanView BIT NOT NULL DEFAULT 0,
    CanAdd BIT NOT NULL DEFAULT 0,
    CanEdit BIT NOT NULL DEFAULT 0,
    CanDelete BIT NOT NULL DEFAULT 0,
    CanExport BIT NOT NULL DEFAULT 0,
    
    -- معلومات التدقيق
    CreatedDate DATETIME2 NOT NULL DEFAULT GETDATE(),
    ModifiedDate DATETIME2 NULL,
    CreatedByUserProfileID INT NULL,
    
    -- العلاقات
    CONSTRAINT FK_Permissions_Screens FOREIGN KEY (ScreenID) 
        REFERENCES Screens(ScreenID),
    CONSTRAINT FK_Permissions_JobTitles FOREIGN KEY (JobTitleID) 
        REFERENCES CustomJobTitle(CustomJobTitleID),
    CONSTRAINT FK_Permissions_Organizations FOREIGN KEY (OrganizationID) 
        REFERENCES Organizations(OrganizationID),
    
    -- منع التكرار: لا يمكن تكرار نفس الشاشة لنفس المسمى في نفس المنظمة
    CONSTRAINT UQ_Permissions_Screen_Job_Org UNIQUE 
        (ScreenID, JobTitleID, OrganizationID)
);

-- فهارس للأداء
CREATE INDEX IX_Permissions_JobTitle_Org ON Permissions(JobTitleID, OrganizationID);
CREATE INDEX IX_Permissions_Screen ON Permissions(ScreenID);
```

### **2.2 تعديل بسيط على MenuItems (لربط القوائم بالشاشات)**

```sql
-- إضافة حقل ScreenID لربط عنصر القائمة بالشاشة
ALTER TABLE MenuItems ADD ScreenID INT NULL;
ALTER TABLE MenuItems ADD CONSTRAINT FK_MenuItems_Screens 
    FOREIGN KEY (ScreenID) REFERENCES Screens(ScreenID);

-- تحديث البيانات الحالية (مثال)
-- UPDATE MenuItems SET ScreenID = 21 WHERE ItemCode = 'DISEASES';
-- UPDATE MenuItems SET ScreenID = 22 WHERE ItemCode = 'PSP_PROGRAMS';
```

### **2.3 تعديل بسيط على Screens (إلغاء MinimumRoleLevel)**

```sql
-- MinimumRoleLevel غير ضروري الآن، يمكن تركه أو إلغاؤه
-- سنعتمد كلياً على Permissions
ALTER TABLE Screens DROP COLUMN MinimumRoleLevel; -- اختياري
```
