## 2.2 BaseViewModel (النمط الآمن)

```csharp
using CommunityToolkit.Mvvm.ComponentModel;
using CommunityToolkit.Mvvm.Input;

namespace RubikCare.Mobile.Infrastructure.ViewModels
{
    public partial class BaseViewModel : ObservableObject
    {
        [ObservableProperty]
        private bool _isLoading;

        [ObservableProperty]
        private string _errorMessage = string.Empty;

        [ObservableProperty]
        private bool _isRefreshing;

        protected readonly IApiService _apiService;
        protected readonly IAuthService _authService;

        public BaseViewModel(IApiService apiService, IAuthService authService)
        {
            _apiService = apiService;
            _authService = authService;
        }

        [RelayCommand]
        private async Task LoadDataAsync()
        {
            await ExecuteAsync(OnLoadDataAsync);
        }

        protected virtual Task OnLoadDataAsync() => Task.CompletedTask;

        protected async Task ExecuteAsync(Func<Task> operation)
        {
            if (IsLoading) return;

            try
            {
                IsLoading = true;
                ErrorMessage = string.Empty;
                await operation();
            }
            catch (HttpRequestException ex)
            {
                ErrorMessage = "فشل الاتصال بالخادم. تأكد من اتصالك بالإنترنت.";
                Console.WriteLine($"Network Error: {ex.Message}");
            }
            catch (UnauthorizedAccessException)
            {
                ErrorMessage = "انتهت صلاحية الجلسة. الرجاء تسجيل الدخول مرة أخرى.";
                await _authService.LogoutAsync();
                await Shell.Current.GoToAsync("//LoginPage");
            }
            catch (Exception ex)
            {
                ErrorMessage = $"حدث خطأ: {ex.Message}";
                Console.WriteLine($"Error: {ex.Message}");
            }
            finally
            {
                IsLoading = false;
                IsRefreshing = false;
            }
        }
    }
}
```

## 2.3 ApiService (التواصل مع الخادم)

```csharp
using System.Text;
using System.Text.Json;

namespace RubikCare.Mobile.Infrastructure.Services
{
    public interface IApiService
    {
        Task<T?> GetAsync<T>(string endpoint);
        Task<T?> PostAsync<T>(string endpoint, object data);
        Task<T?> PutAsync<T>(string endpoint, object data);
        Task<bool> DeleteAsync(string endpoint);
        void SetAuthToken(string token);
        void ClearAuthToken();
    }

    public class ApiService : IApiService
    {
        private readonly HttpClient _httpClient;
        private readonly JsonSerializerOptions _jsonOptions;

        public ApiService()
        {
            var baseUrl = DeviceInfo.Current.Platform == DevicePlatform.Android
                ? "http://10.0.2.2:5235"  // Android Emulator
                : DeviceInfo.Current.Platform == DevicePlatform.iOS
                    ? "http://localhost:5235"  // iOS Simulator
                    : "http://localhost:5235"; // Windows Machine

            _httpClient = new HttpClient
            {
                BaseAddress = new Uri(baseUrl),
                Timeout = TimeSpan.FromSeconds(30)
            };
            _httpClient.DefaultRequestHeaders.Add("Accept", "application/json");

            _jsonOptions = new JsonSerializerOptions
            {
                PropertyNameCaseInsensitive = true,
                PropertyNamingPolicy = JsonNamingPolicy.CamelCase
            };
        }

        public void SetAuthToken(string token)
        {
            _httpClient.DefaultRequestHeaders.Authorization =
                new System.Net.Http.Headers.AuthenticationHeaderValue("Bearer", token);
        }

        public void ClearAuthToken()
        {
            _httpClient.DefaultRequestHeaders.Authorization = null;
        }

        public async Task<T?> GetAsync<T>(string endpoint)
        {
            var response = await _httpClient.GetAsync(endpoint);
            response.EnsureSuccessStatusCode();
            var json = await response.Content.ReadAsStringAsync();
            return JsonSerializer.Deserialize<T>(json, _jsonOptions);
        }

        public async Task<T?> PostAsync<T>(string endpoint, object data)
        {
            var json = JsonSerializer.Serialize(data, _jsonOptions);
            var content = new StringContent(json, Encoding.UTF8, "application/json");
            var response = await _httpClient.PostAsync(endpoint, content);
            response.EnsureSuccessStatusCode();
            var responseJson = await response.Content.ReadAsStringAsync();
            return JsonSerializer.Deserialize<T>(responseJson, _jsonOptions);
        }

        public async Task<T?> PutAsync<T>(string endpoint, object data)
        {
            var json = JsonSerializer.Serialize(data, _jsonOptions);
            var content = new StringContent(json, Encoding.UTF8, "application/json");
            var response = await _httpClient.PutAsync(endpoint, content);
            response.EnsureSuccessStatusCode();
            var responseJson = await response.Content.ReadAsStringAsync();
            return JsonSerializer.Deserialize<T>(responseJson, _jsonOptions);
        }

        public async Task<bool> DeleteAsync(string endpoint)
        {
            var response = await _httpClient.DeleteAsync(endpoint);
            return response.IsSuccessStatusCode;
        }
    }
}
```

## 2.5 التنقل بين XAML و Blazor (النمط المختلط)

| الحالة | التقنية | السبب |
|--------|---------|-------|
| صفحات مستقرة وتعمل | XAML | "ما يعمل لا نلمسه" |

### التنقل من XAML إلى Blazor

```csharp
public interface ILoggerService
{
    void LogInfo(string message);
    void LogError(string message, Exception ex = null);
    void LogWarning(string message);
}

public class LoggerService : ILoggerService
{
    private readonly string _logFilePath;

    public LoggerService()
    {
        _logFilePath = Path.Combine(FileSystem.AppDataDirectory, "logs", $"app_{DateTime.Now:yyyyMMdd}.log");
        Directory.CreateDirectory(Path.GetDirectoryName(_logFilePath));
    }

    public void LogError(string message, Exception ex = null)
    {
        var logEntry = $"[ERROR] {DateTime.Now:yyyy-MM-dd HH:mm:ss} - {message}";
        if (ex != null)
            logEntry += $"\nException: {ex.Message}\nStack Trace: {ex.StackTrace}";
        
        File.AppendAllText(_logFilePath, logEntry + Environment.NewLine);
        
        // في وضع التطوير، اطبع أيضاً في Console
        Debug.WriteLine(logEntry);
    }

    public void LogInfo(string message)
    {
        var logEntry = $"[INFO] {DateTime.Now:yyyy-MM-dd HH:mm:ss} - {message}";
        File.AppendAllText(_logFilePath, logEntry + Environment.NewLine);
        Debug.WriteLine(logEntry);
    }

    public void LogWarning(string message)
    {
        var logEntry = $"[WARN] {DateTime.Now:yyyy-MM-dd HH:mm:ss} - {message}";
        File.AppendAllText(_logFilePath, logEntry + Environment.NewLine);
        Debug.WriteLine(logEntry);
    }
}
```

# الجزء 4: نظام الدعوات والرموز

## 4.1 نظرة عامة على نظام الدعوات

نظام الدعوات في RubikCare يسمح بـ:
- **الطبيب** بدعوة مريض للانضمام لبرنامج دعم (عبر QR Code أو رمز قصير)
- **المندوب** بدعوة طبيب أو صيدلي للانضمام للمنصة
- **المريض** باستخدام رمز الدعوة للانضمام للبرنامج

### تدفق الدعوة الكامل

```
طبيب/مندوب ينشئ دعوة
    ↓
نظام يولد رمزاً فريداً (INV-XXXXXX)
    ↓
رمز الدعوة يظهر كـ QR Code + نص قصير
    ↓
الطبيب/المندوب يشارك الرمز مع المستهدف
    ↓
المستهدف (مريض/طبيب/صيدلي) يدخل الرمز في التطبيق
    ↓
نظام يتحقق من صحة الرمز وصلاحيته
    ↓
إذا كان صالحاً → يسجل المستخدم في البرنامج/المنصة
```

---

## 4.2 شرح الرمز القصير (Short Code)

### تركيب الرمز

```
INV - 5 C W E F H E J
 │    └─────────┬─────────┘
 │              │
 │              └─── رمز عشوائي (8 أحرف)
 │
 └─── بادئة ثابتة (INV = Invitation)
```

### أمثلة من النظام

| الرمز | نوع الدعوة | المستهدف |
|-------|------------|----------|
| `INV-5CWEFHEJ` | دعوة من طبيب | مريض |
| `REF-ABC123` | دعوة من مندوب | طبيب/صيدلي |
| `DR-INF-6952` | دعوة قديمة (تنسيق آخر) | مريض |

### كيف نستخدمه في التطبيق

```csharp
// في InvitePatientViewModel.cs
private async Task<string> GenerateInvitationToken()
{
    // توليد رمز عشوائي مكون من 8 أحرف
    const string chars = "ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789";
    var random = new Random();
    var randomPart = new string(Enumerable.Repeat(chars, 8)
        .Select(s => s[random.Next(s.Length)]).ToArray());
    
    return $"INV-{randomPart}";
}

// عندما يدخل المريض الرمز "INV-5CWEFHEJ"
public async Task<InvitationData> DecodeShortCode(string shortCode)
{
    // مثال: "INV-5CWEFHEJ"
    var parts = shortCode.Split('-');
    if (parts.Length == 2 && parts[0] == "INV")
    {
        var invitationToken = shortCode;
        
        // البحث في قاعدة البيانات عن الدعوة
        return await _apiService.GetAsync<InvitationData>(
            $"api/invitations/validate?token={invitationToken}");
    }
    return null;
}
```

## 4.3 QR Code vs الرمز القصير

| الخاصية | QR Code | الرمز القصير |
|---------|---------|--------------|
| **المحتوى** | رابط كامل (deep link) | رمز مختصر (8-12 حرف) |
| **الطول** | طويل (50+ حرف) | قصير |
| **الاستخدام** | مسح بالكاميرا | كتابة يدوية |
| **سرعة الإدخال** | سريع (مسح) | بطيء (كتابة) |
| **الدقة** | عالية (لا خطأ) | قد يحدث خطأ كتابي |
| **متى يستخدم** | في العيادة (حضورياً) | عن بعد (واتساب، رسالة) |

## 4.5 نظام دعوات المندوب (Rep Invitations)

### سيناريو دعوة طبيب من مندوب

```
مندوب في التطبيق
    ↓
يختار برنامجاً من برامجه
    ↓
يبحث عن طبيب (بالتخصص، المنطقة، الاسم)
    ↓
يختار طريقة التواصل (إيميل، واتساب، نسخ الرابط)
    ↓
يضغط "إرسال دعوة"
    ↓
نظام ينشئ دعوة (SourceType = "REP_TO_DOCTOR")
    ↓
يتم إنشاء رمز فريد (REF-XXX-XXX)
    ↓
يتم إرسال الرابط للطبيب
```

### الرابط الناتج

```
https://test.rubikcare.com/join/REF-ABC123?rep=5&program=2
```

### تتبع الدعوة (للمندوب)

```csharp
// في RepDashboardViewModel.cs
public async Task LoadInvitationsAsync()
{
    var response = await _apiService.GetAsync<List<InvitationDto>>(
        $"api/rep/invitations/track?repId={CurrentRepId}");
    
    Invitations = response;
    
    // إحصائيات
    TotalSent = Invitations.Count;
    TotalAccepted = Invitations.Count(i => i.Status == "ACCEPTED");
    TotalConverted = Invitations.Count(i => i.Status == "CONVERTED");
    ConversionRate = TotalSent > 0 
        ? (double)TotalAccepted / TotalSent * 100 
        : 0;
}
```

### حالات الدعوة للمندوب

| الحالة | المعنى |
|--------|--------|
| `PENDING` | في انتظار الرد |
| `ACCEPTED` | تم قبول الدعوة (سجل في المنصة) |
| `EXPIRED` | انتهت صلاحية الدعوة |
| `CONVERTED` | تم تحقيق هدف (صرف دواء) |

## 4.6 التقارير المطلوبة من جدول الدعوات

### 1. عدد الدعوات الناجحة لكل طبيب

```sql
SELECT 
    u.DisplayName AS DoctorName,
    COUNT(i.InvitationID) AS TotalInvitations,
    SUM(CASE WHEN i.Status = 'ACCEPTED' THEN 1 ELSE 0 END) AS SuccessfulInvitations,
    CAST(SUM(CASE WHEN i.Status = 'ACCEPTED' THEN 1 ELSE 0 END) AS FLOAT) / 
        NULLIF(COUNT(i.InvitationID), 0) * 100 AS ConversionRate
FROM PSPInvitations i
JOIN UserProfiles u ON i.InvitedByUserID = u.UserProfileID
WHERE i.SourceType IN ('QR', 'SMS', 'EMAIL')
  AND i.InvitationDate >= DATEADD(MONTH, -1, GETDATE())
GROUP BY u.DisplayName;
```

### 2. عدد الدعوات لكل مندوب

```sql
SELECT 
    u.DisplayName AS RepName,
    COUNT(i.InvitationID) AS InvitationsSent,
    SUM(CASE WHEN i.Status = 'ACCEPTED' THEN 1 ELSE 0 END) AS Accepted,
    SUM(CASE WHEN i.Status = 'CONVERTED' THEN 1 ELSE 0 END) AS Converted
FROM PSPInvitations i
JOIN UserProfiles u ON i.ReferralUserID = u.UserProfileID
WHERE i.SourceType = 'REP_TO_DOCTOR'
  AND i.InvitationDate >= DATEADD(MONTH, -1, GETDATE())
GROUP BY u.DisplayName;
```

### 3. تتبع رحلة المريض من QR إلى صرف الدواء

```sql
SELECT 
    i.InvitationID,
    i.InvitationToken,
    i.InvitationDate,
    i.ResponseDate,
    i.Status,
    p.PatientID,
    p.EnrollmentDate,
    pd.DispensationID,
    pd.TokenCode,
    pd.TokenUsedDate
FROM PSPInvitations i
LEFT JOIN PSPPatients p ON i.UsedByUserID = p.PatientProfileID
LEFT JOIN PSPeRX e ON p.PatientID = e.PatientID
LEFT JOIN PSPDispensations pd ON e.ERxID = pd.ERxID
WHERE i.SourceType = 'QR'
  AND i.InvitationID = @invitationId;
```

### **3. نمط تسجيل CSS (الموجود والمنظم)**

**الهيكل الحالي (صحيح ولا يحتاج تغيير):**

```
wwwroot/css/
├── main.css                    ← الملف الرئيسي (imports فقط)
├── _core/                      ← الأساسيات
├── _components/                ← المكونات
├── _layout/                    ← التخطيطات
├── _pages/                     ← صفحات لها تنسيق خاص
│   └── Messaging/              ← ⭐ صفحات المراسلة الجديدة
│       ├── doctor-search.css
│       ├── doctor-profile.css
│       └── chat.css
├── _themes/                    ← السمات
└── _utilities/                 ← الأدوات المساعدة
```

**كيفية إضافة ملف CSS جديد:**

1.  **أنشئ الملف** في المجلد المناسب (مثل `_pages/Messaging/chat.css`).
2.  **استورده** في `main.css` أو `_PagesBundle.css`:

```css
/* في _PagesBundle.css */
@import url('_pages/Messaging/doctor-search.css');
@import url('_pages/Messaging/doctor-profile.css');
@import url('_pages/Messaging/chat.css');
```

3.  **في صفحة Blazor، أضف الرابط** (إذا كان الملف غير مستورد في main.css):

```razor
<link href="_content/RubikCare.Shared.UI/css/_pages/Messaging/chat.css" rel="stylesheet" />
```

**ملاحظة:** `_content/RubikCare.Shared.UI/` هو المسار التلقائي للملفات الثابتة في الـ RCL.
