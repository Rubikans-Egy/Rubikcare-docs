# 🔄 نظام الدوائر المغلقة (Closed Loop System):
**(الخطة أ)**
## الدائرة 1: الطبيب ↔ المنصة ↔ المريض

### الطبيب (مزود خدمة طبية):
  ├── يدفع اشتراك للاستفادة من أدوات إدارة العيادة
  ├── يدفع عمولة ثابتة (6 روبيكا) لكل حجز مريض عبر المنصة
  └── يحصل على مرضى أكثر ووقت إداري أقل

### المنصة (ميسر):
  ├── توفر أدوات إدارة احترافية للطبيب
  ├── تصل الطبيب بمرضى جدد
  └── تحصل على اشتراكات + عمولات حجز

### المريض (مستفيد مجاني):
  ├── يحصل على خدمة حجز سهلة ومجانية
  ├── يحصل على سجل طبي إلكتروني مجاني
  └── لا يدفع أي شيء للمنصة (قيمة مجتمعية وأخلاقية)


## الدائرة 2: الصيدلية ↔ المنصة ↔ المريض
text
### الصيدلية (مزود خدمة دوائية):
  ├── تدفع عمولة ثابتة (6 روبيكا) لكل طلب دواء عبر المنصة
  ├── تحصل على عملاء جدد من المرضى المحجوزين
  └── تستفيد من نظام إدارة الطلبات والتوصيل

### المنصة (وسيط):
  ├→ تربط المريض بالصيدلية القريبة
  ├→ تقدم نظام إدارة الطلبات للصيدلية
  └→ تحصل على عمولات توصيل أدوية

### المريض (مستفيد):
  ├→ يحصل على توصيل دواء سهل
  ├→ يجد الصيدليات القريبة منه
  └→ لا يدفع للمنصة (العمولة على الصيدلية فقط)


## الدائرة 3: شركات الأدوية ↔ المنصة ↔ الأطباء ↔ المرضى
text
### شركات الأدوية (راعي برامج الدعم):
  ├→ تدفع للمنصة مقابل إدارة برامج دعم المرضى
  ├→ تصل للأطباء والمرضى المستهدفين
  └→ تزيد مبيعاتها من خلال برامج الدعم المنظمة

### المنصة (منظم وميسر):
  ├→ تقدم Medical CRM للشركات
  ├→ تدير برامج دعم المرضى
  └→ تحصل على عقود سنوية من شركات الأدوية

### الطبيب (شريك في النشر):
  ├→ يحصل على أدوات إضافية مجانية
  ├→ يحصل على مرضى من برامج الدعم
  └→ يزيد دخله من صرف الأدوية المدعمة

### المريض (مستفيد رئيسي):
  ├→ يحصل على أدوية مدعومة أو مجانية
  ├→ يحصل على متابعة طبية منتظمة
  └→ لا يدفع للمنصة (التمويل من شركات الأدوية)

## مصادر الدخل الرئيسية لمنصة Rubikcare:

1.  **اشتراكات إدارة العيادات**
    *   **العميل:** الطبيب أو العيادة
    *   **القيمة:** تبدأ من 400 إلى 1500 روبيكا
    *   **الدفع:** شهرياً
    *   **مقابل ماذا؟** استخدام أدوات متكاملة لإدارة العيادة تشمل: جدولة المواعيد، وإدارة سجلات المرضى، ونظام الفواتير.

2.  **عمولات الحجز عبر الإنترنت**
    *   **العميل:** الطبيب
    *   **القيمة:** 6 روبيكا
    *   **الدفع:** مقابل كل حجز
    *   **مقابل ماذا؟** عمولة ثابتة عن كل موعد يحجزه مريض عبر المنصة.

3.  **عمولات توصيل الأدوية**
    *   **العميل:** الصيدلية
    *   **القيمة:** 6 روبيكا
    *   **الدفع:** مقابل كل طلب
    *   **مقابل ماذا؟** عمولة ثابتة عن كل طلب دواء يتم توصيله عبر خدمة المنصة.

4.  **عقود برامج دعم المرضى**
    *   **العميل:** شركات الأدوية
    *   **القيمة:** 10,000 روبيكا فما فوق
    *   **الدفع:** سنوياً
    *   **مقابل ماذا؟** توفير وإدارة برامج دعم متابعة المرضى، بالإضافة إلى نظام إدارة علاقات العملاء الطبي (Medical CRM).

5.  **الإعلانات المستهدفة**
    *   **العميل:** مقدمي الخدمات الطبية والتجهيزات
    *   **القيمة:** متغيرة
    *   **الدفع:** حسب الحملة الإعلانية
    *   **مقابل ماذا؟** عرض إعلانات مُخصصة داخل التطبيق لمقدمي الخدمات المهتمين بالوصول إلى جمهورنا.

---

### ما نرفض تقاضي أي رسوم عنه مطلقاً:

لضمان شفافيتنا وتركيزنا على المريض، **لا نأخذ أبداً**:
*   **أي رسوم من المرضى** مقابل استخدام الخدمات الأساسية.
*   **نسبة أو عمولة** من كشوفات الأطباء أو قيمة الاستشارات.
*   **رسوم على كتابة أو إصدار الوصفات الطبية**.
*   **عمولات على سعر الأدوية نفسها** (عمولتنا الوحيدة مرتبطة بتكلفة *التوصيل* فقط).

## **نموذج الأعمال: كيف نخلق قيمة للشركاء ونحقق الاستدامة؟**

تعمل منصتنا على نموذج متعدد الأطراف، حيث نربط بين مختلف مقدمي الخدمات الصحية لخلق تجربة متكاملة. وفيما يلي تفصيل للعلاقة الاقتصادية والقيمة المقدمة لكل شريك:

#### **1. الأطباء والعيادات (العملاء الأساسيون)**
*   **طبيعة العلاقة:** **شركاء و عملاء مدفوعون**.
*   **التدفق النقدي للمنصة:** يشمل **اشتراكًا شهريًا** (يتراوح بين 400 و1500 روبيكا) **+ عمولة ثابتة قدرها 6 روبيكا** عن كل حجز 

#### مريض عبر المنصة.
*   **القيمة التي يحصلون عليها:**
    *   **نظام إدارة عيادة رقمي متكامل:** يشمل جدولة المواعيد، وإدارة السجلات الطبية للمرضى، ونظام الفواتير والمدفوعات الإلكترونية.
    *   **قناة تسويقية فعالة:** الوصول إلى قاعدة واسعة من المرضى الباحثين عن خدماتهم وزيادة معدلات الحجز.
    *   **إنتاجية أعلى:** تطبيق موبايل مخصص للإدارة، مع إشعارات وتذكيرات تلقائية لتحسين سير العمل وتقليل المواعيد الفائتة.
    *   **تكامل سلس:** مع خدمات الصيدليات لتسهيل كتابة الوصفات الإلكترونية (eRX) وتتبعها.

#### **2. الصيدليات (شركاء في سلسلة التوريد)**
*   **طبيعة العلاقة:** **عملاء مدفوعون**.
*   **التدفق النقدي للمنصة:** **عمولة ثابتة قدرها 6 روبيكا** عن كل طلب دواء يتم توصيله عبر المنصة.
*   **القيمة التي تحصل عليها:**
    *   **منصة طلبات وتوصيل مخصصة:** نظام كامل لإدارة طلبات الأدوية من الاستلام إلى التسليم، مع تتبع في الوقت الفعلي.
    *   **مصدر تدفق مرضى مضمون:** وصول مباشر إلى مرضى الأطباء المشتركين في الشبكة، مما يزيد من حجم المبيعات.
    *   **أدوات ذكاء الأعمال:** ميزات تساعد في إدارة المخزون وتحليل اتجاهات المبيعات.
    *   **تكامل عميق:** مع منصات الأطباء لاستقبال الوصفات الإلكترونية (eRX) بشكل آلي وفوري، مما يقلل الأخطاء ويسرع الخدمة.

#### **3. شركات الأدوية (شركاء استراتيجيون)**
*   **طبيعة العلاقة:** **عملاء بمنتجات وخدمات مخصصة (B2B)**.
*   **التدفق النقدي للمنصة:** **عقود سنوية بقيمة تبدأ من 10,000 روبيكا**، حسب نطاق الخدمة.
*   **القيمة التي تحصل عليها:**
    *   **منصة Medical CRM متطورة:** أداة متكاملة لإدارة العلاقات مع المهنيين الصحيين ومرضى أمراض محددة.
    *   **إدارة برامج دعم المرضى:** حلول رقمية كاملة لتقديم الدواء، والمتابعة، والتوعية، وقياس الالتزام بالعلاج.
    *   **وصول دقيق وقياسي:** القدرة على استهداف جماهير محددة من الأطباء والمرضى ضمن معايير ديموغرافية وتشخيصية معينة.
    *   **تقارير تحليلية مفصلة:** رؤى قابلة للقياس حول فعالية البرامج، وسلوك المرضى، وأثر المبادرات العلاجية.

### **التدفقات النقدية المترابطة: دائرة القيمة**

يمثل هذا الهيكل **دائرة قيمة مغلقة**:
1.  **الطبيب** يدفع اشتراكًا لتحسين كفاءة عيادته وجذب المرضى.
2.  **المريض** يحجز مجانًا، ويتلقى الرعاية، ويتم توجيه وصفتِه بسلاسة.
3.  **الصيدلية** تدفع عمولة توصيل صغيرة مقابل طلب جديد تم توصيله إليه عبر النظام.
4.  **شركة الدواء** تستثمر في المنصة للتواصل مع هذا النظام البيئي الصحي بكفاءة وقياس النتائج.
5.  **العائدات** تُعاد استثمارها في تطوير المنصة والحفاظ على **مجانيتها الكاملة للمرضى** وسلامة تعاملاتهم.

هذا النموذج المتوازن هو أساس التزامنا:
 **تمكين مقدمي الخدمات بمنتجات مميزة، مع حماية المريض من أي رسوم خفية أو تضارب في المصالح.**

public class PSPRevenueModel
{
    // 1. رسوم إعداد البرنامج
    public decimal SetupFee { get; set; } = 5000; // 5,000 روبيكا
    
    // 2. رسوم إدارة شهري
    public decimal MonthlyManagementFee { get; set; } = 1000; // 1,000 روبيكا/شهر
    
    // 3. عمولة على كل مريض
    public decimal PerPatientCommission { get; set; } = 50; // 50 روبيكا/مريض
    
    // 4. عقد سنوي مع الشركة
    public decimal AnnualContractValue(int estimatedPatients)
    {
        return SetupFee + 
               (MonthlyManagementFee * 12) + 
               (PerPatientCommission * estimatedPatients);
    }
    
    // مثال: برنامج يستهدف 200 مريض
    // = 5000 + (1000×12) + (50×200) = 5000 + 12000 + 10000 = 27,000 روبيكا
    // = 27,000 × 5 = 135,000 جنيه مصري سنوياً
}
