## 4. نظام الألوان والخلفيات - المبدأ الأساسي

### ⚠️ المشكلة التي حُلّت
كانت `_typography.css` تحتوي على هذه القاعدة الخطرة:
```css
/* ❌ خاطئ - يُجبر اللون الداكن على كل عنصر */
body, p, span, div:not([class*="btn"]) {
    color: var(--rubik-text-primary);
}
```
هذه القاعدة كانت تمنع توريث الألوان وتُسبب ظهور النصوص داكنة
حتى على الخلفيات الداكنة.

### ✅ الحل المعتمد
```css
/* ✅ صحيح - body فقط، الباقي يرث */
body {
    color: var(--rubik-text-primary);
}
```

### المبدأ:
```
خلفية فاتحة (افتراضي) → نصوص داكنة عبر التوريث من body
خلفية داكنة           → أضف class="... rubik-dark"
```

### استخدام `.rubik-dark`
أضف هذه الـ class على **أي عنصر خلفيته داكنة**:

```html
<!-- ✅ hero sections -->
<section class="dashboard-hero rubik-dark">...</section>

<!-- ✅ card headers الداكنة -->
<div class="card-header rubik-dark">
    <h6>إحصائيات طبية</h6>
</div>

<!-- ✅ brand في الـ topbar -->
<div class="ch-topbar-brand rubik-dark">
    <span>RubikCare</span>
</div>

<!-- ✅ أي div بخلفية داكنة -->
<div class="bg-primary rubik-dark">...</div>
```

**النتيجة:** كل النصوص داخله تصبح بيضاء تلقائياً بدون أي CSS إضافي.

### استثناءات `.rubik-dark`
```css
/* الأزرار تحافظ على ألوانها */
.rubik-dark .btn { color: unset; }

/* ألوان الحالات لا تُلغى */
.rubik-dark .text-warning { color: var(--rubik-warning) !important; }
.rubik-dark .text-success { color: var(--rubik-success) !important; }
.rubik-dark .text-danger  { color: var(--rubik-danger)  !important; }
.rubik-dark .text-muted   { color: rgba(255,255,255,.65) !important; }
```

### 6.2 نمط تسمية مفاتيح الترجمة ResourceKey

```
DOMAIN.PAGE.TYPE.KEY

مثال:
PSP.PROGRAM_EDIT.TITLE.NEW
PSP.PROGRAM_EDIT.BUTTON.SAVE
PSP.PROGRAM_EDIT.LABEL.PROGRAM_NAME
PSP.PROGRAM_EDIT.MESSAGE.SAVE_SUCCESS
PSP.PROGRAM_EDIT.EMPTY_STATE.NO_SPECIALTIES
```

**المجالات (Domains):**
```
COMMON        ← مشترك بين الكل
ADMIN         ← لوحة التحكم
PSP           ← برامج دعم المرضى
PHARMA        ← بوابة شركات الأدوية
CLINIC        ← بوابة العيادات
PHARMACY      ← بوابة الصيدليات
PATIENT       ← بوابة المرضى
EXTERNAL_LOGIN← صفحة تسجيل الدخول الخارجي
CENTERHUB     ← الصفحة الرئيسية
```

### 6.3 تحميل الترجمات في الصفحات

**صفحات Interactive (InteractiveServer):**
```csharp
@inject TranslationStateService TranslationState
@inject LocalizationService Localizer
@implements IDisposable

@code {
    private string _currentLang = "ar";
    private Dictionary<string, string> _translations = new();

    protected override async Task OnInitializedAsync()
    {
        _currentLang = TranslationState.CurrentLanguage ?? "ar";
        TranslationState.OnLanguageChanged += HandleLanguageChanged;
        await LoadTranslations(_currentLang);
    }

    private async void HandleLanguageChanged()
    {
        _currentLang = TranslationState.CurrentLanguage ?? "ar";
        await LoadTranslations(_currentLang);
        await InvokeAsync(StateHasChanged);
    }

    private async Task LoadTranslations(string lang)
    {
        var pageTrans   = await Localizer.GetPageTranslationsAsync("PAGE_DOMAIN", lang);
        var commonTrans = await Localizer.GetPageTranslationsAsync("COMMON", lang);
        _translations.Clear();
        foreach (var item in commonTrans) _translations[item.Key] = item.Value;
        foreach (var item in pageTrans)   _translations[item.Key] = item.Value;
    }

    private string T(string key) =>
        _translations.TryGetValue(key, out var v) ? v : key;

    public void Dispose() =>
        TranslationState.OnLanguageChanged -= HandleLanguageChanged;
}
```

**صفحات Static (بدون InteractiveServer):**
```csharp
// قراءة اللغة من Query String أو Cookie
var queryLang  = HttpContext.Request.Query["lang"].ToString();
var cookieLang = HttpContext.Request.Cookies["lang"];
var currentLang = !string.IsNullOrEmpty(queryLang)  ? queryLang  :
                  !string.IsNullOrEmpty(cookieLang) ? cookieLang : "ar";

// حقن الترجمات في JavaScript
<script>
    window.__initialTranslations = @Html.Raw(JsonSerializer.Serialize(
        _translations,
        new JsonSerializerOptions {
            Encoder = JavaScriptEncoder.UnsafeRelaxedJsonEscaping
        }));
    window.__currentLang = "@currentLang";
</script>
```

### 6.4 تحديث الترجمات عبر JavaScript (Static Pages)

```javascript
// login-localization.js
const loginLocalization = {
    updateTranslations: async function(lang) {
        // جلب من domains متعددة دفعة واحدة
        const [loginRes, commonRes, externalRes] = await Promise.all([
            fetch(`/api/localization/page/LOGIN?lang=${lang}`),
            fetch(`/api/localization/page/COMMON?lang=${lang}`),
            fetch(`/api/localization/page/EXTERNAL_LOGIN?lang=${lang}`)
        ]);
        const translations = {
            ...await commonRes.json(),
            ...await loginRes.json(),
            ...await externalRes.json()
        };
        // تحديث DOM
        document.querySelectorAll('[data-translate]').forEach(el => {
            const key = el.getAttribute('data-translate');
            // البحث case-insensitive
            const value = translations[key]
                       || translations[key.toUpperCase()]
                       || translations[key.toLowerCase()];
            if (value) el.textContent = value;
        });
        document.documentElement.dir  = lang === 'ar' ? 'rtl' : 'ltr';
        document.documentElement.lang = lang;
        localStorage.setItem('RubikCare:Language', lang);
        document.cookie = `lang=${lang}; path=/; max-age=31536000; SameSite=Lax`;
    }
};

window.toggleLanguageSimple = function() {
    const current = document.documentElement.lang || 'ar';
    loginLocalization.updateTranslations(current === 'ar' ? 'en' : 'ar');
};
```

## 10. دليل المطور

### للمطور الجديد - ابدأ هنا

```
1. اقرأ _core/_variables.css لتعرف المتغيرات المتاحة
2. استخدم .rubik-dark على أي عنصر بخلفية داكنة
3. لا تضع CSS داخل ملفات Razor
4. اختبر في RTL (عربي) و LTR (إنجليزي) دائماً
5. تحقق من DevTools → Computed عند أي تعارض ألوان
```

### عند إضافة صفحة جديدة

```
1. أنشئ _pages/_pagename.css
2. أضف @import في main.css في القسم 7
3. استخدم prefix فريد لكل الـ classes (.pn-)
4. أضف سجل قاعدة بيانات للـ translations بـ DOMAIN الجديد
5. وثّق الصفحة في جدول 8.1
```

### عند تعارض الألوان - خطوات التشخيص

```
1. افتح DevTools (F12) → انقر على العنصر
2. تبويب Computed → ابحث عن color
3. انظر من أي ملف تأتي القاعدة المطبقة
4. إذا كانت من _typography.css → تأكد من حذف قاعدة span/div القديمة
5. إذا كانت الخلفية داكنة → أضف .rubik-dark على العنصر
6. إذا كانت من Bootstrap → أضف .rubik-dark أو override في _components
```
