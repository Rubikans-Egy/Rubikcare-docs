# نموذج المنصة الاقتصادي
## Platform Economic Model

---

### فلسفة المنصة: الدوائر المغلقة الثلاثة

روبيك كير ليست تطبيقاً واحداً، بل **نظام بيئي متعدد الأطراف** يعتمد على ثلاث دوائر اقتصادية مغلقة، كل منها تغذي الأخرى:

```
┌─────────────────────────────────────────────────────────┐
│                    الدائرة الأولى                        │
│              الطبيب ↔ المنصة ↔ المريض                    │
│  (إدارة العيادة - الحجز - السجل الطبي)                   │
├─────────────────────────────────────────────────────────┤
│                    الدائرة الثانية                       │
│             الصيدلية ↔ المنصة ↔ المريض                  │
│     (الوصفات الإلكترونية - طلب الدواء - التوصيل)        │
├─────────────────────────────────────────────────────────┤
│                    الدائرة الثالثة                       │
│    شركة الأدوية ↔ المنصة ↔ الطبيب ↔ المريض              │
│        (برامج دعم المرضى - الالتزام بالعلاج)            │
└─────────────────────────────────────────────────────────┘
```

---

### الدائرة الأولى: الطبيب ↔ المنصة ↔ المريض

**من يدفع:** الطبيب (اشتراك + عمولة)
**من يستفيد مجاناً:** المريض

| العنصر | التفاصيل | القيمة |
|--------|----------|--------|
| **اشتراك العيادة** | 400 - 1,500 روبيكا/شهر | أدوات إدارة متكاملة |
| **عمولة الحجز** | 6 روبيكا لكل حجز | مريض جديد عبر المنصة |
| **القيمة للطبيب** | مرضى أكثر، وقت إداري أقل، سمعة رقمية | زيادة الدخل 20-30% |
| **القيمة للمريض** | حجز سهل، سجل طبي، تذكير بالمواعيد | صحة منظمة مجاناً |

**الاقتصاديات:**
- CAC (تكلفة اكتساب الطبيب): ~50 روبيكا (عبر مندوبي الدعاية)
- LTV (قيمة العميل مدى الحياة): 2,400 روبيكا (متوسط 20 شهر × 120 روبيكا)
- **LTV/CAC = 48:1** (نسبة استثنائية)

---

### الدائرة الثانية: الصيدلية ↔ المنصة ↔ المريض

**من يدفع:** الصيدلية (عمولة توصيل)
**من يستفيد مجاناً:** المريض (التوصيل مدعوم من الصيدلية)

| العنصر | التفاصيل | القيمة |
|--------|----------|--------|
| **عمولة التوصيل** | 6 روبيكا لكل طلب | طلب دواء عبر المنصة |
| **القيمة للصيدلية** | عملاء جدد، إدارة طلبات، تتبع مخزون | زيادة المبيعات 15-25% |
| **القيمة للمريض** | توصيل للمنزل، مقارنة أسعار، وصفات رقمية | راحة وتوفير وقت |

**الاقتصاديات:**
- CAC (تكلفة اكتساب الصيدلية): ~30 روبيكا (عبر شبكة الأطباء)
- LTV: 1,800 روبيكا (متوسط 25 طلب/شهر × 12 شهر × 6 روبيكا)
- **LTV/CAC = 60:1**

---

### الدائرة الثالثة: شركة الأدوية ↔ المنصة ↔ الطبيب ↔ المريض

**من يدفع:** شركة الأدوية (نسبة من المبيعات)
**من يستفيد:** الطبيب (أدوات إضافية)، المريض (دواء مدعوم)

| العنصر | التفاصيل | القيمة |
|--------|----------|--------|
| **نموذج الدخل** | 3% من مبيعات برنامج الدعم | PSP منظم وقابل للقياس |
| **القيمة للشركة** | التحكم في الخصومات، منع حرق الأسعار، بيانات الالتزام | زيادة الربحية 5-15% |
| **القيمة للطبيب** | أدوات PSP مجانية، مرضى من البرامج، متابعة سهلة | دخل إضافي من العلاجات المدعومة |
| **القيمة للمريض** | دواء مخفض أو مجاني، متابعة منتظمة | التزام بالعلاج |

**الاقتصاديات:**
- CAC (تكلفة اكتساب الشركة): ~5,000 روبيكا (بيع مباشر عبر العلاقات)
- LTV: 27,000+ روبيكا (برنامج 200 مريض × متوسط 4,500 روبيكا/مريض × 3%)
- **LTV/CAC = 5.4:1** (مع تكرار العقود سنوياً)

---

### نظام الروبيكا: العملة الداخلية

**الروبيكا (Rubik):** وحدة خدمية داخلية تجمع بين:
1. **وحدة دفع:** 1 روبيكا = 5 جنيه مصري (ثابتة)
2. **نقاط تحفيز:** مكافآت للأنشطة الصحية الإيجابية
3. **وسيط تبادل:** تحويلات بين المستخدمين داخل المنصة

**الدورة الاقتصادية:**

```
الأطباء/الصيدليات/الشركات ← يشترون روبيكا بالجنيه
         ↓
المنصة ← تحتفظ بالفائض كإيراد
         ↓
توزيع روبيكا مجانية ← مكافآت للمرضى والأطباء النشطين
         ↓
التبادل الداخلي ← المريض يدفع للطبيب/الصيدلية مقابل خصومات
         ↓
إعادة الدورة ← الطبيب/الصيدلية يستخدمون الروبيكا للدفع للمنصة
```

---

### تأثيرات الشبكة (Network Effects)

**1. تأثير الشبكة المباشر (Direct Network Effects):**
كلما زاد عدد المرضى في المنصة، زادت قيمتها لكل مريض جديد (بيانات أكثر، أطباء أكثر).

**2. تأثير الشبكة المتقاطع (Cross-Side Network Effects):**
- المزيد من الأطباء ← يجذب المزيد من المرضى ← يجذب المزيد من الصيدليات ← يجذب المزيد من شركات الأدوية

**3. تأثير البيانات (Data Network Effects):**
كل مريض جديد يحسن خوارزميات الذكاء الاصطناعي للمنصة (تشخيص مساعد، توصيات علاجية).

**عجلة النمو الذاتية (Growth Flywheel):**

```
شركات الأدوية ← تدفع للمنصة
      ↓
مندوبو الدعاية ← يدعون الأطباء
      ↓
الأطباء ← يدخلون المرضى للبرامج
      ↓
المرضى ← يتفاعلون مع الصيدليات
      ↓
الشبكة تكبر ← تجذب المزيد من شركات الأدوية
```

---

### Unit Economics التفصيلية

| المؤشر | الطبيب | الصيدلية | شركة الأدوية | المريض |
|--------|--------|----------|--------------|--------|
| **CAC** | 50 روبيكا | 30 روبيكا | 5,000 روبيكا | 0 (مجاني) |
| **ARPU/Month** | 120 روبيكا | 150 روبيكا | 2,250 روبيكا* | 0 |
| **Churn Rate** | 5% | 8% | 10% | 15% |
| **LTV** | 2,400 روبيكا | 1,800 روبيكا | 27,000 روبيكا | N/A |
| **LTV/CAC** | 48:1 | 60:1 | 5.4:1 | ∞ |
| **Payback Period** | 0.4 شهر | 0.2 شهر | 2.2 شهر | فوري |

*متوسط برنامج 200 مريض × 4,500 روبيكا/مريض × 3% ÷ 12 شهر

## Section 2: Platform Economic Model

### The Rubik Token Economy

**Rubik (RUB):** Triple-function internal unit:
1. **Service Unit:** 1 RUB = 5 EGP (fixed)
2. **Reward Points:** Incentivizing healthy behaviors
3. **Exchange Medium:** Internal transactions

**Why tokens?** Creates circular economy, reduces friction, enables micro-transactions, and builds platform loyalty.

---

### Unit Economics by Stakeholder

| Metric | Doctors | Pharmacies | Pharma Companies | Patients |
|--------|---------|------------|------------------|----------|
| **CAC** | 50 RUB | 30 RUB | 5,000 RUB | Free |
| **Monthly Revenue** | 120 RUB | 150 RUB | 2,250 RUB* | $0 |
| **Churn** | 5% | 8% | 10% | 15% |
| **LTV** | 2,400 RUB | 1,800 RUB | 27,000 RUB | N/A |
| **LTV/CAC** | **48:1** | **60:1** | **5.4:1** | Infinite |
| **Payback** | 0.4 months | 0.2 months | 2.2 months | Immediate |

*Average 200-patient program × 4,500 RUB/patient × 3% ÷ 12 months

**Platform-wide LTV/CAC: 38:1** (weighted average)

---

### Network Effects Flywheel

```
Pharma Companies (pay) → Medical Reps (distribute) 
                                ↓
Doctors (enroll) → Patients (engage) → Pharmacies (fulfill)
                                ↓
                    Network grows → More pharma companies
```

**Data Network Effects:** Each patient improves AI diagnosis assistance, treatment recommendations, and adherence prediction for future patients.
