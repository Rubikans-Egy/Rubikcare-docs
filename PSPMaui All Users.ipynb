# 1️⃣ **رحلة المندوب (Pharma Rep)**

## 📋 **وصف المستخدم**
مندوب مبيعات يعمل لدى شركة أدوية، مسؤول عن:
- الترويج لبرامج الدعم لدى الأطباء والصيادلة
- متابعة انضمام الأطباء للبرامج
- تحقيق أهداف مبيعات

# كيفية تفعيل تتبع الرموز القصيرة عند إرسال دعوات

## 📝 **شرح الرمز القصير (Short Code)**

الرمز القصير `DR-INF-6952` هو **مفتاح مختصر** يمثل بيانات كاملة. دعني أشرح لك كيف يعمل:

### **تركيب الرمز:**
```
DR  -  INF  -  6952
 │      │        │
 │      │        └─── الرقم العشوائي (لتمييز كل دعوة)
 │      │
 │      └─────────── كود البرنامج المختصر (INF = Infertility)
 │
 └────────────────── معرف الطبيب (DR = Doctor)
```

### **كيف نستخدمه؟**

عندما يدخل المريض هذا الرمز في التطبيق:

```csharp
// في InvitePatientViewModel.cs
private async Task<string> GetProgramCode(string programName)
{
    return programName switch
    {
        "Infertility Support Program" => "INF",
        "AnemiaFree" => "ANF",
        "HeartCare Plus" => "HCP",
        _ => programName.Length > 3 ? programName.Substring(0, 3).ToUpper() : "PRG"
    };
}

// عندما يدخل المريض الرمز "DR-INF-6952"
public async Task<InvitationData> DecodeShortCode(string shortCode)
{
    // مثال: "DR-INF-6952"
    var parts = shortCode.Split('-');
    if (parts.Length == 3 && parts[0] == "DR")
    {
        var programCode = parts[1];  // "INF"
        var randomCode = parts[2];    // "6952"
        
        // البحث في قاعدة البيانات عن:
        // 1. البرنامج الذي له هذا الكود المختصر
        // 2. الطبيب (المفترض أننا نعرفه من سياق الجلسة)
        // 3. العيادة (نعرفها من الطبيب)
        
        return await _apiService.GetAsync<InvitationData>($"api/invitations/decode?code={programCode}&random={randomCode}");
    }
    return null;
}
```

## 🗄️ **هيكل قاعدة البيانات المقترح للرموز القصيرة**

```sql
CREATE TABLE ShortCodes (
    ShortCodeID INT PRIMARY KEY IDENTITY(1,1),
    ShortCode NVARCHAR(20) NOT NULL,           -- DR-INF-6952
    ProgramID INT NOT NULL,                     -- 2
    ClinicID INT NOT NULL,                       -- 1245
    DoctorProfileID INT NOT NULL,                -- 12
    RandomCode INT NOT NULL,                      -- 6952
    IsUsed BIT DEFAULT 0,
    ExpiryDate DATETIME2,
    CreatedDate DATETIME2 DEFAULT GETDATE()
);

-- عند إنشاء رمز جديد
INSERT INTO ShortCodes (ShortCode, ProgramID, ClinicID, DoctorProfileID, RandomCode)
VALUES ('DR-INF-6952', 2, 1245, 12, 6952);

-- عند استخدام الرمز
SELECT * FROM ShortCodes 
WHERE ShortCode = 'DR-INF-6952' 
  AND IsUsed = 0 
  AND (ExpiryDate IS NULL OR ExpiryDate > GETDATE());
```

## 🔄 **الفرق بين QR Code والرمز القصير**

| الخاصية | QR Code | الرمز القصير |
|---------|---------|--------------|
| **المحتوى** | رابط كامل (deep link) | رمز مختصر |
| **الطول** | طويل (50+ حرف) | قصير (10-12 حرف) |
| **الاستخدام** | مسح بالكاميرا | كتابة يدوية |
| **سرعة الإدخال** | سريع (مسح) | بطيء (كتابة) |
| **الدقة** | عالية (لا خطأ) | قد يحدث خطأ كتابي |
| **الانترنت** | يحتاج لربط | يمكن كتابته بدون نت ثم استخدامه لاحقاً |

## ✅ **كيف نربط الرمز القصير بالبيانات في التطبيق؟**

عندما يدخل المريض الرمز `DR-INF-6952`:

1. التطبيق يرسل الرمز للـ API: `POST /api/invitations/validate`
2. الـ API يبحث في جدول `ShortCodes` عن هذا الرمز
3. إذا وجده، يرجع بيانات البرنامج والعيادة والطبيب
4. التطبيق يفتح صفحة التسجيل مع هذه البيانات مسبقة التعبئة

```csharp
// مثال للـ API endpoint
[HttpPost("validate")]
public async Task<IActionResult> ValidateShortCode([FromBody] ValidateCodeRequest request)
{
    var code = await _context.ShortCodes
        .FirstOrDefaultAsync(c => c.ShortCode == request.ShortCode 
                                && !c.IsUsed 
                                && (c.ExpiryDate == null || c.ExpiryDate > DateTime.UtcNow));
    
    if (code == null)
        return NotFound("رمز غير صالح أو منتهي الصلاحية");
    
    var program = await _context.PSPPrograms.FindAsync(code.ProgramID);
    var clinic = await _context.Organizations.FindAsync(code.ClinicID);
    
    return Ok(new 
    {
        ProgramId = program.ProgramID,
        ProgramName = program.ProgramNameAr,
        ClinicName = clinic.OrganizationNameAr,
        DoctorId = code.DoctorProfileID
    });
}
```

## 📈 **التقارير المطلوبة - كلها ممكنة**

### **1. عدد الدعوات الناجحة لكل طبيب**

```sql
SELECT 
    u.DisplayName AS DoctorName,
    COUNT(i.InvitationID) AS TotalInvitations,
    SUM(CASE WHEN i.Status = 'ACCEPTED' THEN 1 ELSE 0 END) AS SuccessfulInvitations,
    CAST(SUM(CASE WHEN i.Status = 'ACCEPTED' THEN 1 ELSE 0 END) AS FLOAT) / 
        NULLIF(COUNT(i.InvitationID), 0) * 100 AS ConversionRate
FROM PSPInvitations i
JOIN UserProfiles u ON i.InvitedByUserID = u.UserProfileID
WHERE i.SourceType = 'QR'
  AND i.InvitationDate >= DATEADD(MONTH, -1, GETDATE())
GROUP BY u.DisplayName;
```

### **2. عدد الدعوات لكل مندوب (عند دعوة طبيب)**

```sql
SELECT 
    u.DisplayName AS RepName,
    COUNT(i.InvitationID) AS InvitationsSent,
    SUM(CASE WHEN i.Status = 'ACCEPTED' THEN 1 ELSE 0 END) AS Accepted
FROM PSPInvitations i
JOIN UserProfiles u ON i.ReferralUserID = u.UserProfileID
WHERE i.SourceType = 'REP_TO_DOCTOR'
  AND i.InvitationDate >= DATEADD(MONTH, -1, GETDATE())
GROUP BY u.DisplayName;
```

### **3. تتبع رحلة المريض من QR إلى صرف الدواء**

```sql
SELECT 
    i.InvitationID,
    i.InvitationDate,
    i.ResponseDate,
    i.Status,
    p.ParticipationID AS PatientEnrollment,
    pd.DispensationID AS FirstDispensation
FROM PSPInvitations i
LEFT JOIN PSPParticipations p ON i.ResultingParticipationParticipationID = p.ParticipationID
LEFT JOIN PSPDispensations pd ON p.ParticipationID = pd.ParticipationID
WHERE i.SourceType = 'QR'
  AND i.InvitationID = @invitationId;
```

## 📘 **الفصل الجديد: نظام برامج دعم المرضى (PSP)**

## 🌿 **8. نظام برامج دعم المرضى (Patient Support Programs - PSP)**

### **8.1 نظرة عامة**

نظام PSP في RubikCare يسمح للأطباء بدعوة مرضاهم للاستفادة من برامج الدعم الدوائي التي تقدمها شركات الأدوية.

**الدورة الكاملة:**
```
1. شركة أدوية → تنشئ برنامج دعم (PSPProgram)
2. طبيب → يشترك في البرنامج (PSPParticipation)
3. طبيب → ينشئ دعوة لمريض (PSPInvitation)
4. مريض → يدخل كود الدعوة في التطبيق
5. نظام → يسجل المريض وينشئ وصفة ورمز صرف
6. مريض → يستخدم رمز الصرف في الصيدلية
```

### **8.3 APIs**

#### **8.3.1 إنشاء دعوة - `create-invitation`**

**Endpoint:** `POST /api/psp/create-invitation`

**Request:**
```json
{
    "programId": 2,
    "clinicId": 1245
}
```

**Response:**
```json
{
    "success": true,
    "invitationId": 12,
    "invitationToken": "INV-5CWEFHEJ",
    "message": "تم إنشاء الدعوة بنجاح"
}
```

**المنطق:**
1. التحقق من أن الطبيب مشترك في البرنامج
2. توليد رمز فريد (INV-XXXXXX)
3. إنشاء سجل في `PSPInvitations` (Status = PENDING)
4. إرجاع الرمز للطبيب

#### **8.3.2 نقطة دخول المريض - `entry`**

**Endpoint:** `POST /api/psp/entry`

**Request:**
```json
{
    "invitationCode": "INV-5CWEFHEJ"
}
```
