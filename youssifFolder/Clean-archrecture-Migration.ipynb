## 📚 **الدروس المستفادة والمشاكل التي واجهتنا**

### **1. مشكلة تحويل `object` إلى `DbContext`** 🔴
الخطأ: CS1061 'object' does not contain a definition for 'Set<T>'
السبب: IDbContextFactoryService كان يعيد object بدلاً من DbContext
الحل: إضافة تحويل (DbContext)context في كل دالة قبل استخدام الـ Set

### **2. مشكلة فقدان الـ Navigation Properties** 🔴
الخطأ: CS1061 'TEntity' does not contain a definition for 'CategoryNameAr'
السبب: الخدمة الجديدة لم تكن تقوم بـ Include للعلاقات تلقائياً
الحل: إضافة Include لكل العلاقات في GetAllAsync أو استخدام IncludeRelatedEntities في الصفحات

### **3. مشكلة تضارب الأسماء (Ambiguous Reference)** 🟡
الخطأ: CS0104 'GenericService<>' is an ambiguous reference
السبب: وجود الخدمة القديمة والجديدة بنفس الاسم في مشروعين مختلفين
الحل: استخدام الـ namespace الكامل عند التسجيل في Program.cs مؤقتاً

### **4. مشكلة حقن الخدمات (Injection)** 🟡
الخطأ: Cannot provide a value for property 'Localizer'
السبب: الصفحات كانت تحقن الكلاس مباشرة (LocalizationService) بدلاً من الواجهة (ILocalizationService)
الحل: تغيير جميع الصفحات لاستخدام الواجهة

### **5. مشكلة الدوال المفقودة في الواجهات** 🟡
الخطأ: CS0535 'LocalizationService' does not implement interface member
السبب: إضافة دوال للواجهة دون إضافتها للكلاس
الحل: التأكد من تطابق الواجهة والكلاس دائماً

### **6. مشكلة Async/Await مع Add** 🔴
الخطأ: CS1061 مع AddAsync
السبب: استخدام Add بدلاً من AddAsync مع await
الحل: استخدام await dbContext.Set<T>().AddAsync(entity)

## ✅ **الحل النهائي**

### **تعديل DesignTimeDbContextFactory ليصبح مستقلاً:**

```csharp
using Microsoft.EntityFrameworkCore;
using Microsoft.EntityFrameworkCore.Design;

namespace RubikCare.Infrastructure
{
    public class DesignTimeDbContextFactory : IDesignTimeDbContextFactory<BusinessDbContext>
    {
        public BusinessDbContext CreateDbContext(string[] args)
        {
            var optionsBuilder = new DbContextOptionsBuilder<BusinessDbContext>();
            
            // ✅ ConnectionString مباشرة (مستقلة عن Api.Web)
            var connectionString = "Server=WIN-T0JRQNO9KMA\\RUBICKCARE;Database=RubikCare_Dev_V2;Trusted_Connection=True;TrustServerCertificate=True;MultipleActiveResultSets=true";
            
            optionsBuilder.UseSqlServer(connectionString);
            return new BusinessDbContext(optionsBuilder.Options);
        }
    }
}
```

### **النتيجة بعد التعديل:**
- ✅ EF Core يعمل بشكل مستقر

#### **4. تحديث DynamicMenuService:**

```csharp
private async Task<List<MenuItemDto>> GetOrganizationSpecificMenusAsync(int organizationId)
{
    return await _dbFactory.ExecuteReadOnlyAsync(async context =>
    {
        var dbContext = (DbContext)context;
        
        var assignments = await dbContext.Set<MenuAssignment>()
            .Include(ma => ma.SystemMenu)
                .ThenInclude(m => m.Items)
            .Where(ma => ma.OrganizationID == organizationId && ma.IsActive)
            .ToListAsync();

        var items = new List<MenuItemDto>();
        foreach (var assignment in assignments)
        {
            items.AddRange(assignment.SystemMenu.Items
                .Where(i => i.IsVisible)
                .Select(i => new MenuItemDto
                {
                    ItemID = i.ItemID,
                    ItemNameAr = i.ItemNameAr ?? string.Empty,
                    Url = i.Url?.Replace("{OrganizationID}", organizationId.ToString()) ?? string.Empty,
                    IconClass = i.IconClass ?? string.Empty,
                    SortOrder = i.SortOrder,
                    OrganizationId = organizationId
                }));
        }

        return items.OrderBy(i => i.SortOrder).ToList();
    });
}
```

### **وظائف SQL المطلوبة:**

```sql
-- 1. إنشاء نسخة احتياطية
CREATE PROCEDURE sp_CreateBackup
    @BackupName NVARCHAR(200),
    @Description NVARCHAR(500) = NULL,
    @CreatedBy NVARCHAR(100) = NULL
AS
BEGIN
    DECLARE @BackupPath NVARCHAR(500);
    DECLARE @Timestamp NVARCHAR(20) = FORMAT(GETDATE(), 'yyyyMMdd_HHmmss');
    
    SET @BackupPath = 'C:\Backup\RubikCare_' + @Timestamp + '.bak';
    
    BACKUP DATABASE [RubikCare_Dev_V2] 
    TO DISK = @BackupPath
    WITH FORMAT, INIT, 
         NAME = @BackupName,
         DESCRIPTION = @Description,
         STATS = 10;
    
    INSERT INTO BackupHistory (BackupPath, BackupName, BackupSizeMB, DatabaseName, CreatedBy, Description)
    VALUES (@BackupPath, @BackupName, 
            (SELECT CAST(backup_size / 1048576 AS DECIMAL(10,2)) 
             FROM msdb.dbo.backupset 
             WHERE database_name = 'RubikCare_Dev_V2' 
             ORDER BY backup_start_date DESC),
            'RubikCare_Dev_V2', @CreatedBy, @Description);
    
    SELECT SCOPE_IDENTITY() AS BackupID;
END
```
