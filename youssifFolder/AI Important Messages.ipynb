### Create ORGANIZATION
عايزك ترجع تبعتها تاني بعد ما تخليها احترافية زي الصفحات السابقة كده يا سيد يا حبيبي
مع إضافة محلوظة مهمة جدا جدا جدا مش عايزك تغير اي حاجة في الترجمة ولا نظام الترجمة ولا ال direction
=====================
وعايزك تتناقش معايا في المحتوى وايه اللي ممكن نضيفه او نعدله بص 
الصفحة دي المفروض هتفتح لكل المستخدمين بتوع البرنامج 
وهنخلي المنظمات اللي المستخدم يقدر ينشأها 3:
شركة أدوية 
صيدلية
عيادة
في الوقت الحالي هنكتفي بال3 دول وعايزين بقا نخلي ان المستخدم العادي يكون ظاهرله ينشيء عيادة طلاما هو لسه مش متسجل في السيستم انه دكتور مثلا أو صيدلي 
ونخلي في مثلا زرار options 
ده يظهرله 
(فبالتالي بعدها تتعمل بطريقة تسلسلية ان يظهرله انه يعمل صيدلية )أنا صيدلي
(فبالتالي بعدها تتعمل بطريقة تسلسلية ان يظهرله انه يعمل عيادة )أنا دكتور 
وهنخلي الصيدلية او العيادة دي مش مفعلة او بمعنى تاني مش verfied 
وبالتالي في السيستم مش هنظهرها أونلاين ولا هيكون جمبها علامة ال verficiation أو الصح دي 
فدلوقتي بقا قولي التصور الكامل بتاعك بعد اللي انت فهمته وعايزين نتناقش في المحتوى بتاع الصفحة هل مثلا لانك برضه ممكن تقترتح حاجات مقولتهاش


الصورة واضحة 100% دلوقتي يا يوسف!

---

**الـ 2 Verifications المستقلتين:**

```
Verification 1: الشخص نفسه       Verification 2: المنظمة
─────────────────────────────    ──────────────────────────
"أنا دكتور موثق"                 "عيادتي موثقة"
    ↕ مستقلة تماماً                   ↕ مستقلة تماماً
```

**الـ 4 سيناريوهات الممكنة:**

| الشخص | المنظمة | النتيجة |
|-------|---------|---------|
| ✅ دكتور موثق | ✅ عيادة موثقة | كل حاجة شغالة — ظاهر للعموم |
| ✅ دكتور موثق | ❌ عيادة غير موثقة | هو موثق كدكتور بس العيادة local فقط |
| ❌ غير موثق كدكتور | ✅ عيادة موثقة | نادر — ممكن يحصل لو Admin وثق العيادة بشكل منفرد |
| ❌ غير موثق | ❌ غير موثقة | تجريبي كامل — local only |

---

**سؤالي الأخير قبل الكود:**

لما المستخدم بيكون في صفحة إنشاء المنظمة وبيختار "أنا دكتور" — هل في نفس اللحظة دي بيبدأ طلب توثيق الشخص تلقائياً؟ ولا ده flow منفصل خالص في صفحة تانية؟

ممتاز! الصورة كاملة دلوقتي. خليني أعمل الصفحة:الـ Flow النهائي:

اختار النوع → لو عيادة/صيدلية ظهرله optional prompt "هل تريد طلب التوثيق الشخصي الآن؟" مع زرار Skip → يكمل الـ form عادي
شركة أدوية → مباشرة للـ form بدون أي prompt
منع التكرار من نفس النوع

### Organization SETTINGS (Members Titels)

بص بقا يا سيد يا حبيبي انا مبسوط منك بسبب شغل امبارح النهاردة بقا عايزين نعمل حاجة تانية 
مهمة جداااا 
تاريخ النهاردة 3/17/2026 التلات
===========================
Controll level : ده مستويات التحكم عموما ومش مربوطة بال job titles 
ولا نفسها (صلاحيات الشخص جوا البرنامج)
===================
لكن ال job titles دي بتبقى المسميات الوظيفية فثط في كل منظمة 
(دوو الشخص في شركته)
==================
ال permissions: 
1)Access control (control levels)
2)Screens
3)Assign users to control levels
يعني كده ال control levels جزء من moduleال permissions

ده شرح بعض العنواين اللي هذكرها دلوقتي بس عشان تبقى فاهم الفرق 
بص يا سيدي 
في سيكشن المنظمات ليهم تاب اسمها OrgSettings
 الصفحة دي بتفتح متقسمة ل 
4 أو 5 صفحات بنتنقل فيهم في نفس الصفحة اللي تعتبر ال HUB
وهما :
* Orgainzation Info
* Job Titels (المسميات الوظيفية داخل الشركة)
* Members (إدارة الأعضاء)
* Members Titels (تعريفات الأعضاء داخل المنظمة بمسمياتهم الوظيفية)
* Permissions
* Controll Levels 
ده الشكل القديم الجديد اننا هنفصل ال permission ل tab لوحدها في السيكشن بتاع المنظمات عشان هي صفحة كبيرة وفيها شغل كتير وبرضه ال Controll Levels هتبقى معاها لأن يعتبر ال controll levels زي sub من ال permissions
فاللي عايزين نوصله حاليا في صحة ال Org Settings (HUB)
يكون دول الللي عندنا 

* Orgainzation Info
* Job Titels (المسميات الوظيفية داخل الشركة)
* Members (إدارة الأعضاء)
* Members Titels (متعملتش لسخ ودي اللي هنشتغل عليها دلوقتي 
وبرضه ملهاش جدول في قاعدة البيانات فعايزين نعملها مع العلم ان البرنامج هنا Code First )

ف اللي عايز نبدأ بيه بقا اننا نعمل استعلامات في ال DataBase 
عشان تفهم الشكل والتربيطات اللي هنعملها 
واطلع على الفايلات وال decumentation اللي بعتهالك عشان تفهم ال migration هنعمله ازاي 

ده كده شرح للوضع الحالي قولي كده اللي فهمته عشان نكون متفقين عليه ولو عايز تفهم او تسأل اي حاجة قولي على طول
